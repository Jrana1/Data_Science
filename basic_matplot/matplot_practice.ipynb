# Load Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#resize the graph
plt.figure(figsize=(10,8),dpi=100)
#line1
plt.plot(x,y,'b^--',label='2x')
# line2
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title('Our First Graph',fontdict={'fontsize':20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

# Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]
plt.figure(figsize=(5,5),dpi=100)
bars = plt.bar(labels,values)
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
```

# Load fifa Data and inspect

```python
fifa_data=pd.read_csv('fifa_data.csv')
fifa_data.head(10)
```

## Histogram

```python
bins=[40,50,60,70,80,90,100]
plt.figure(figsize=(8,5))
plt.hist(fifa_data.Overall,bins=bins,color='blue')
plt.xticks(bins)
plt.xlabel("Number of player")
plt.ylabel("Skill level")
plt.title('Distribution of players skills in FIFA 2018')
plt.show()
```

# Pie Chart

```python
df = fifa_data
left = df.loc[df['Preferred Foot']=='Left'].count()[0]
right = df.loc[df['Preferred Foot']=='Right'].count()[0]
plt.figure(figsize=(8,5))
labels=['left','right']
colors = ['blue','green']
plt.pie([left,right],labels=[label.upper() for label in labels],colors=colors,autopct="%.2f%%")
plt.show()
```

# Box plot

```python
barcelona = df.loc[ df.Club=='FC Barcelona']['Overall']
madrid = df.loc[ df.Club=='Real Madrid']['Overall']
revs = df.loc[ df.Club=='New England Revolution']['Overall']
bp=plt.boxplot([barcelona,madrid,revs],labels=['Barcelona','Real Madrid','NE Revolution'],vert=False)
plt.title("Professional Soccer Team Comparison")
for box in bp['boxes']:
    box.set(color='green',linewidth=2)
 
plt.show()
```
