```python
from graphviz import Digraph
from IPython.display import display

# Create a binary tree using graphviz
dot = Digraph()
dot.node('A', 'Hello world')
dot.node('B', 'B')
dot.node('C', 'C')
dot.node('D', 'D')
dot.node('E', 'E')
dot.node('F', 'F')
dot.edges(['AB', 'BC', 'CD', 'DE', 'EF'])

# Render the binary tree in the notebook
display(dot)
```

```python
import sympy as sp

# Define a symbolic variable
x = sp.Symbol('x')

# Define a mathematical expression
expr = x**2 + 2*x + 1

# Differentiate the expression
deriv = sp.diff(expr, x)

# Print the result
print("Derivative of the expression:", deriv)
```
